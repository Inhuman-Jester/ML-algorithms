```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Randomly generate an array of 100 nubers for X
X = np.random.rand(100, 1)

# Made a Y with relation of m=2 and c=3
Y = 2 * X + 3
```

```python
#Function for estimated Y
def predicted(m, X, c):
    Y_pred = m * X + c
    return Y_pred
```

```python
# Loss Function
def loss(Y, m, X, c):
    Y_pred = predicted(m, X, c)
    loss = np.mean((Y-Y_pred) ** 2)
    return loss
```

```python
# This function returns the gradient wrt m and c
def gradient(Y, m, X, c):
    Y_pred = predicted(m, X, c)
    d_m = -2 * np.mean(X * (Y-Y_pred))
    d_c = -2 * np.mean(Y-Y_pred)
    return d_m, d_c
```

```python
# This function performs gradient descent
def gradDescent(iterations, learn_rate, Y, m, X, c):
    
    #storing values of d_m, d_c and loss to see the trend later on
    del_m = []
    del_c = []
    loss_func = []

    #iterate multiple times to get hhigher accuracy
    while iterations != 0:
        d_m, d_c = gradient(Y, m, X, c)
        
        m -= learn_rate * d_m
        c -= learn_rate * d_c
        
        del_m.append(m)
        del_c.append(c)
        loss_func.append(loss(Y, m, X, c))
        
        iterations -= 1
        
    return m, c, del_m, del_c, loss_func
```

```python
# Train model defining learning rate and no. of iterations
iterations = 5000
learn_rate = 0.01

m = 0
c = 0

m, c, del_m, del_c, loss_func = gradDescent(iterations, learn_rate, Y, m, X, c)
```

```python
plt.scatter(X, Y)
plt.plot(X, predicted(m, X, c), color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.show
```

```python
plt.plot(del_m)
plt.xlabel('No. of iterations')
plt.ylabel('Value of m')
plt.title('Variation in value of Slope (m)')
plt.show
```

```python
plt.plot(del_c)
plt.xlabel('No. of iterations')
plt.ylabel('Value of c')
plt.title('Variation in value of Intercept (c)')
plt.show
```

```python
plt.plot(loss_func)
plt.title('Loss Function')
plt.xlabel('no. of iteration')
plt.ylabel('Loss/Error Value')
```
